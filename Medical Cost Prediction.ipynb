# Medical Cost Prediction

### Objective of the Analysis

The goal of this analysis is to predict the medical expenses of patients based on their demographic and lifestyle information. By understanding the factors that influence medical costs, we can gain insights into healthcare expenses and identify key variables that contribute to higher costs. This analysis uses the **Insurance Dataset** from [Kaggle](https://www.kaggle.com/mirichoi0218/insurance).

### Dataset Overview

The dataset contains **1338 observations** and **7 variables**, which include both numerical and categorical data. Below is a detailed description of the variables:

| Variable   | Description                                                                 |
|------------|-----------------------------------------------------------------------------|
| **age**    | Age of the primary beneficiary (in years).                                 |
| **sex**    | Gender of the beneficiary (`male` or `female`).                            |
| **bmi**    | Body Mass Index (BMI), a measure of body fat based on height and weight.   |
| **children** | Number of children/dependents covered by the health insurance.            |
| **smoker** | Smoking status of the beneficiary (`yes` for smoker, `no` for non-smoker). |
| **region** | Residential area of the beneficiary in the US (`northeast`, `northwest`, `southeast`, `southwest`). |
| **charges** | Individual medical costs billed by health insurance (in USD).             |

This dataset provides a comprehensive view of patient demographics, lifestyle habits, and their corresponding medical expenses, making it suitable for predictive modeling and exploratory data analysis.

```python
#importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('insurance.csv')
df.head()
```

## Data Preprocessing

```python
#number of rows and columns
df.shape
```

```python
#checking for missing values
df.info()
```

```python
#checking discriptive statistics
df.describe()
```

```python
#value counts for categorical variables
print(df.sex.value_counts(),'\n',df.smoker.value_counts(),'\n',df.region.value_counts())
```

Replacing the categorical variables with numerical values.
- sex : 1 - male, 0 - female
- smoker : 1 - yes, 0 - no
- region : 0 - northeast, 1 - northwest, 2 - southeast, 3 - southwest

```python
#changing categorical variables to numerical
df['sex'] = df['sex'].map({'male':1,'female':0})
df['smoker'] = df['smoker'].map({'yes':1,'no':0})
df['region'] = df['region'].map({'southwest':0,'southeast':1,'northwest':2,'northeast':3})
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df.drop('charges',axis=1), df['charges'], test_size=0.2, random_state=0)
```

## Model Building

### Linear Regression

```python
#Linear Regression
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr
```

```python
#model training
lr.fit(x_train,y_train)
#model accuracy
lr.score(x_train,y_train)
```

```python
#model prediction
y_pred = lr.predict(x_test)
```

### Polynomial Regression

```python
plr = LinearRegression()
#model training
plr.fit(x_train,y_train)
#model accuracy
plr.score(x_train,y_train)
```

```python
#model prediction
y_pred = plr.predict(x_test)
```

### Decision Tree Regressor

```python
#decision tree regressor
from sklearn.tree import DecisionTreeRegressor
dtree = DecisionTreeRegressor()
dtree
```

```python
#model training
dtree.fit(x_train,y_train)
#model accuracy
dtree.score(x_train,y_train)
```

```python
#model prediction
dtree_pred = dtree.predict(x_test)
```

### Random Forest Regressor

```python
#random forest regressor
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators=100)
rf
```

```python
#model training
rf.fit(x_train,y_train)
#model accuracy
rf.score(x_train,y_train)
```

```python
#model prediction
rf_pred = rf.predict(x_test)
```

## Model Evaluation

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

### Linear Regression

```python
#distribution of actual and predicted values
plt.figure(figsize=(7,5))
ax1 = sns.distplot(y_test,hist=False,color='r',label='Actual Value')
sns.distplot(y_pred,hist=False,color='b',label='Predicted Value',ax=ax1)
plt.title('Actual vs Predicted Values for Linear Regression')
plt.xlabel('Medical Expense')
plt.show()
```

```python
print('MAE:', mean_absolute_error(y_test, y_pred))
print('MSE:', mean_squared_error(y_test, y_pred))
print('RMSE:', np.sqrt(mean_squared_error(y_test, y_pred)))
print('R2 Score:', r2_score(y_test, y_pred))
```
